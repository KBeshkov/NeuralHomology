```python
n_stim = 200
n_neurons = 500
n_pieces = 3
mfld_gen = Manifold_Generator()
S1 = mfld_gen(n_stim,'S1',1)
stim_angles = np.linspace(0,2*np.pi,n_stim)

metric_used = geodesic_knn(k=4,adaptive=True).fit
Homologizer = Persistent_Homology()

projection = np.random.randn(n_neurons,2)@S1
def projection_fn(x,iters):
    for i in range(iters):
        x = np.tanh(1.5*np.random.randn(n_neurons,n_neurons)@x)
    x = 0.01*np.random.randn(n_neurons,n_neurons)@x
    return x
tun_curves =  projection_fn(projection,3)
#tun_curves = tun_curves[:,np.random.permutation(n_stim)]
#for i in range(n_neurons):
#    tun_curves[i] = np.polyval(np.polyfit(stim_angles, tun_curves[i],deg=18),stim_angles)
```

```python
plt.plot(stim_angles,tun_curves[5:9].T)
plt.xlabel('angle')
```

```python
pvalue_hom_curves = Homologizer.perm_test(tun_curves.T,metric_used,n_perms=10,pval=100)
hom_curves = Homologizer(tun_curves.T,metric_used,False,0,[1,None])
plot_diagrams(hom_curves[1])#Homologizer.barcode_plot(hom_curves[1],2,pval=pvalue_hom_curves[0])
```

```python
kappas = np.logspace(0,4,20)
neuron_nums = np.logspace(1,4,10).astype(int)#[5,10,25,50,100,250,500,1000]
signif_homs = np.zeros([len(kappas),len(neuron_nums)])

for i, n_nrn in enumerate(neuron_nums):
    tuning_means = np.linspace(-np.pi, np.pi, n_nrn)#np.pi*np.random.randn(n_nrn)#
    for j, kap in enumerate(kappas):
        tun_curves = np.zeros([n_nrn,n_stim])
        for k in range(n_nrn):
            tun_curves[k,:] = stats.vonmises(loc=tuning_means[k],kappa=kap).pdf(stim_angles)
        if i==0:
            plt.plot(tun_curves[0])
        hom_pval = Homologizer.perm_test(tun_curves.T,metric_used,n_perms=10,pval=100)
        signif_homs[j,i] = sum(hom_pval[1] > hom_pval[0])
        
```

```python
plt.imshow(signif_homs==1,aspect='auto',origin='lower')
plt.xticks(np.arange(0,len(neuron_nums)),neuron_nums)
plt.yticks(np.arange(0,len(kappas)),np.round(np.log10(kappas),1))
plt.xlabel('# neurons')
plt.ylabel('$\log_{10}(\kappa)$')
```

```python
tunning_means = np.linspace(-np.pi, np.pi, n_neurons)
opp_angle = np.pi
orth_tunning_neurons = np.zeros([n_neurons,n_stim])
standard_neurons = np.zeros([n_neurons, n_stim])
kappa = 100
for i in range(n_neurons):
    orth_tunning_neurons[i,:] = (stats.vonmises(loc=tunning_means[i],kappa=kappa).pdf(stim_angles)+stats.vonmises(loc=(tunning_means[i]+opp_angle)%(2*np.pi),kappa=kappa).pdf(stim_angles))/2
    standard_neurons[i,:] = stats.vonmises(loc=tunning_means[i],kappa=kappa).pdf(stim_angles)/2
    #orth_tunning_neurons += 0.5*np.random.rand(n_neurons,n_stim)
orth_tunning_neurons = np.vstack([orth_tunning_neurons, standard_neurons])
```

```python
plt.plot(stim_angles,orth_tunning_neurons[np.random.randint(0,2*n_neurons,10)].T)
plt.xlabel('angle')
```

```python
pvalue_hom = Homologizer.perm_test(orth_tunning_neurons.T,metric_used,n_perms=10,pval=100)
hom_data = Homologizer(orth_tunning_neurons.T,metric_used,False,0,[1,None])
cycle_lengths = hom_data[1][1][:,1]-hom_data[1][1][:,0]
```

```python
plt.imshow(hom_data[0])
```

```python
reducer = umap.UMAP(n_neighbors=50,min_dist=0,n_components=2,metric='precomputed')
ori_mfld = reducer.fit_transform(hom_data[0])
```

```python
max_cycle = np.argsort(cycle_lengths).astype(int)[-1]
print('Max cycle index is: ' + str(max_cycle))
dmult=0.1
hom_crossection = hom_data[1][1][max_cycle][0]+ dmult*(hom_data[1][1][max_cycle][1]-hom_data[1][1][max_cycle][0])

Homologizer.barcode_plot(hom_data[1],2,pval=pvalue_hom[0])
```

```python
edge_dist = np.copy(hom_data[0])
edge_dist[edge_dist>hom_crossection] = 0

BA = Barcode_Analyzer()

plt.figure(dpi=300)
plt.subplot(121)
BA.plotCocycle2D(hom_data[0],ori_mfld[:,:2],hom_crossection,labels=stim_angles,node_cmap='seismic')
plt.subplot(122)
plt.imshow(edge_dist, extent=[0,360,360,0])
plt.yticks([])
plt.tight_layout()
#plt.savefig('Nmanifold_euclid.png')
```

```python
#Rotating star
from matplotlib import path

def rot_mat(angle):
    return np.array([[np.cos(angle), np.sin(angle)],[-np.sin(angle),np.cos(angle)]])

n_points = 1000
f_res = 100
k_rays = 5
Star = mfld_gen.Star(n_points,k_rays,2)
mesh = np.meshgrid(np.linspace(-1,1,f_res),np.linspace(-1,1,f_res))
mesh_array = np.vstack([mesh[0].flatten(),mesh[1].flatten()])
p = path.Path(Star.T)  # square with legs length 1 and bottom left corner at the origin

f_star = p.contains_points(mesh_array.T)
```

```python
plt.pcolormesh(np.linspace(-1,1,f_res),np.linspace(-1,1,f_res),f_star.reshape(f_res,f_res),cmap='gray',alpha=0.5)
plt.plot(Star[0],Star[1],'k',lw=3)
```

```python
rotations = np.linspace(0, 2*np.pi, 100)
Stars = np.zeros([len(rotations),Star.shape[0],Star.shape[1]])
f_stars = np.zeros([len(rotations),f_res*f_res])
for i,r in enumerate(rotations):
    Stars[i] = rot_mat(r)@Star
    p = path.Path(Stars[i].T)
    f_stars[i] = p.contains_points(mesh_array.T)
    
```
